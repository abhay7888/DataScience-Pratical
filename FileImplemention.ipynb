```python


# Display the original Package column
print("Original Package column:")
print(df_cleaned['Package'])

# Convert the Package column to string to avoid errors during replacement
df_cleaned['Package'] = df_cleaned['Package'].astype(str)

# Remove specific words, symbols, and unwanted characters from the Package column
df_cleaned['Package'] = df_cleaned['Package'].str.replace(r'Avg|mentioned on offer letter|[\(\),-]', '', regex=True)  # Remove specific words and symbols
df_cleaned['Package'] = df_cleaned['Package'].str.replace(r'[^0-9]', '', regex=True)  # Keep only digits

# Convert the cleaned column to numeric (if applicable)
df_cleaned['Package'] = pd.to_numeric(df_cleaned['Package'], errors='coerce')

# Display the cleaned Package column
print("\nCleaned Package column:")
print(df_cleaned['Package'])
```

# 5].Implement Linear Regression algoritham on given data set

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Drop rows with missing values
df_cleaned = df_cleaned.dropna()

# Select features and target variable (we are predicting 'Percentage')
X = df_cleaned[['Stud_id', 'Package', 'Passing Year']]  # Features
y = df_cleaned['Percentage']  # Target variable

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Linear Regression model
linear_regression = LinearRegression()

# Train the model on the training set
linear_regression.fit(X_train, y_train)

# Make predictions on the test set
y_pred = linear_regression.predict(X_test)

# Evaluate the model's performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Output the results
print("Mean Squared Error:", mse)
print("R² Score:", r2)
```

# 6].Implement Logistic Regression algorithm on a given dataset ?

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

# Create a new target column "Performance" by binning the "Percentage" column into categories
df_cleaned['Performance'] = pd.cut(df_cleaned['Percentage'], bins=[0, 70, 85, 100], labels=['Low', 'Medium', 'High'])

# Select features and target variable
X = df_cleaned[['Stud_id', 'Package', 'Passing Year']]  # Features/coulmns
y = df_cleaned['Performance']  # Target variable (Performance categories)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Logistic Regression model
logistic_regression = LogisticRegression(max_iter=1000)

# Train the model on the training set
logistic_regression.fit(X_train, y_train)

# Make predictions on the test set
y_pred = logistic_regression.predict(X_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Output the results
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", class_report)
```
